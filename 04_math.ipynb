```python
import numpy as np
import pandas as pd
from numpy.linalg import norm
# Import everything I need

titanic = pd.read_csv("Titanic_features.csv")
# Load the dataset
target = titanic["Survived"].values
mean = np.sum(target)/len(target)
print("Mean = ", mean)
std = np.sqrt(np.sum((target-mean)**2)/len(target))
print("Std = ", std)
# Compute the mean and std of target using numpy
X = titanic["Fare"].values
meanX = np.mean(X)
stdX = np.std(X)
Z = (X-meanX)/stdX
print("Z = ",Z[:5])
# Standardise the column fare
high = titanic.loc[titanic["Fare"].idxmax(),["Age","Fare"]].values
low = titanic.loc[titanic["Fare"].idxmin(), ["Age","Fare"]].values
cosine = np.dot(high,low)/(norm(high)*norm(low))
print("Cosine = ", cosine)
# Compute cosine similarity
high_fare = titanic[titanic["Fare"] > titanic["Fare"].mean()]
prob = high_fare["Survived"].mean()
print("Probability of survival for high fare:",prob)
# Probability of survival for high fare
```
